```python
import numpy as np
import matplotlib.pyplot as plt

# 假设有3个隐藏状态
num_states = 3

# 假设有10个时间步骤
num_timesteps = 10

# 生成随机的显著图数据（0到1之间的随机值）
saliency_maps = np.random.rand(num_states, num_timesteps)

# 设置颜色映射
cmap = 'hot'

# 创建一个子图网格，每行显示一个隐藏状态的显著图
fig, axes = plt.subplots(num_states, 1, figsize=(6, 4*num_states))

# 循环遍历每个隐藏状态
for i in range(num_states):
    # 在当前子图中绘制显著图
    axes[i].imshow([saliency_maps[i]], cmap=cmap, aspect='auto')
    
    # 设置子图标题和轴标签
    axes[i].set_title('Hidden State {}'.format(i+1))
    axes[i].set_xlabel('Time Step')
    axes[i].set_ylabel('Saliency')
    
# 调整子图之间的间距
plt.tight_layout()

# 显示图形
plt.show()
```
